```python
# 서울역의 시간대별로 승차 인원 선 그래프를 그리세요
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#가산점
df = pd.read_excel('2019년 05월  교통카드 통계자료.xls',
                   sheet_name=3, encoding='cp949',index_col=3, skiprows=1)
df.shape
```

```python

data = df.iloc[:, 3:-1]
```

```python

data = data.replace(',', '', regex=True) # , 없애기
```

```python

data = data.astype(np.int) # 정수로 변환
```

```python
#서울역이 여러개다 더하자
seoul = data.loc['서울역']
plt.plot(np.sum(seoul[::2], axis=0)) #짝수 : 승차
plt.plot(np.sum(seoul[1::2], axis=0)) #홀수 : 하차
```

- 아침 9~10시 사이에 하차인원이 가장 많은 역 5개 찾기

```python
col = 11
morning_10_off = data.iloc[:, col]
```

```python
y= morning_10_off.sort_values(ascending=False)[:5]
y
```

```python
plt.bar(y.index, y)
```

- 승차인원에 비해서 하차 인원의 비율이 높은 역 5개

```python
data.head()
승차 = data.iloc[:, ::2]
하차 = data.iloc[:, 1::2]
```

```python
승차합 = np.sum(승차, axis=1)
하차합 = np.sum(하차, axis=1)
비율 = 하차합/승차합
```

```python
비율.sort_values(ascending=False)[:5]
```

- 하루 중 승차 인원이 가장 많은 시간은 언제?
- -> 시간대별 합, 최대값

```python
승하차인원 = np.sum(data, axis=0)
승차인원 = 승하차인원[::2]
4 + np.argmax(승차인원.values) # 18시
```

- 무임 하차 비율이 가장 높은 역 5개
- 무임 하차 비율 산출, 역별 정렬, 5개 뽑기

```python
df = pd.read_excel('2019년 05월  교통카드 통계자료.xls',
                   sheet_name=2, encoding='cp949', index_col=3, skiprows=0)
df.shape
```

```python
subway = df.iloc[:, 3:-1]
subway.head()
```

```python
subway = subway.replace(',', '', regex=True)
subway = subway.astype('int')
```

```python
subway.head()
```

```python
유임하차 = subway['유임하차']
무임하차 = subway['무임하차']
# 유임하차
비율 = 무임하차/유임하차
```

```python
y = 비율.sort_values(ascending=False)[:5]
```

```python
m = plt.pie(y, labels=y.index, autopct='%d')
```
