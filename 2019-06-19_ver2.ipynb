- p85~86 처럼 8월과 1월의 최고기온 히스토그램을 그리시오

```python
import csv
f = open('seoul.csv')
f
data = csv.reader(f)
next(data)
result_01 = [] # 비어 있는 리스트 준비
result_08 = [] # 비어 있는 리스트 준비
for row in data : # 한줄씩 꺼낸다
    if row[-1] !='': # 누락되어 있지 않으면
        date = row[0].split('-')
        if date[1] == '01':
            result_01.append(float(row[-1])) # 최고기온을 실수로 변환해서 저장
        elif date[1] == '08': # 6월 19일 100년간 기록은?
            result_08.append(float(row[-1])) # 최고 기온만 추가
```

##### 1~12 의 최고기온을 박스 그림으로 표현합니다. p91~92

```python
import pandas as pd
data = pd.read_csv('seoul.csv', encoding='cp949')
data.head() # 상위 5건
data = data.dropna() # 전처리 : 누락된 값이 있으면 그 행은 버린다.
data.head() # 상위 5건
```

```python
import numpy as np
month = []
for i in range(12):
    month.append([])
x = np.array(data)
x.shape, x[0]
```

```python
np.mean(x[:,-1]), len(x)
```

```python
for i in range(len(x)):
    row = x[i]
    date = row[0]
    high = row[-1]
    mon = date.split('-')[1]
    #1월은 0번째 위치에 저장해야 하므로 -1
    month[int(mon)-1].append(high)
    #print(row)
```

```python
import matplotlib.pyplot as plt
# plt.figure(figsize=(5,1,1))
m = plt.boxplot(month)
```
