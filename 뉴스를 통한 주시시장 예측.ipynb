뉴스를 통해 주식시장 예측

```python
import pandas as pd
import numpy as np
```

```python
folder = 'stocknews/'
files = ['Combined_News_DJIA.csv', 'DJIA_table.csv', 'RedditNews.csv']
```

```python
combined_news = pd.read_csv(folder + files[0])
stock = pd.read_csv(folder + files[1])
reddit = pd.read_csv(folder + files[2])
```

```python
stock.head(2)
```

```python
combined_news = combined_news.set_index('Date')
```

```python
combined_news = combined_news.dropna() # 누락 제거
```

```python
combined_news.isna().sum().sum()
```

```python
sum(combined_news.index < '2014-12-31'), sum(combined_news.index > '2014-12-31')
```

```python
combined_news.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer # 갯수를 세어서 배열로 만든다
```

```python
X = combined_news.drop(columns='Label')
y = combined_news['Label']
combined_news.shape, X.shape, y.shape
```

```python
x_flat = X.values.reshape([-1])
x_flat.shape
```

```python
# 3% 이하 / 97% 이상 등장하는 단어는 제외, 최대 어휘 20만 제한, N그램 2(단어 2개씩 자르기)
vect = TfidfVectorizer(min_df=0.03, max_df=0.97, max_features = 200000, ngram_range=(2,2)).fit(x_flat)
```

```python

vect = CountVectorizer().fit(x_flat) # 문장의 단어들을 쪼개서 어휘사전을 만듬
```

```python
vect
```

```python
len(vect.vocabulary_), vect.vocabulary_ # 어휘 사전 보기
```

```python
x_sparse =  vect.transform(x_flat) # 자연어를 벡터로 변환
```

```python
feature_n = X.shape[1]
train_m = sum(combined_news.index <= '2014-12-31')
feature_n, train_m
```

```python
y_train = y.values[:train_m]
y_test = y.values[train_m:]
x_train = x_sparse[:train_m*feature_n] # 1608일 * 25개의 뉴스
x_test = x_sparse[train_m*feature_n:]
```

```python
y_train.shape, x_train.shape
```

선형 모델, X의 dim이 2이어야 합니다
> 방안 1 : 뉴스 1~25를 합친다
> 방안 2 : sparse_matrix 25개를 합친다(더한다)

```python
len(X)
```

```python
x_agr = []
for i in range(len(X)):
    one_day_news = x_sparse[i*25 : (i+1)*25] # <25x34682>
    new_sum = np.sum(one_day_news, axis=0)  # <34682>
    x_agr.append(new_sum)
len(x_agr)
```

```python
x_2d = np.concatenate(x_agr, axis=0)
x_2d.shape, y.shape # 이차원 배열 <- 선형모델의 X값이 이차원이여야 함
```

```python
y_train = y.values[:train_m]
y_test = y.values[train_m:]
x_train = x_2d[:train_m] # 1608일 * 25개의 뉴스
x_test = x_2d[train_m:]
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
model = RidgeClassifier().fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

주가 데이터(수치형) 6-7개 컬럼 + 뉴스(TEXT > vector) 2-3만개
컬럼 / open, close, low, high, volume, 그날의 뉴스 벡터

주가 데이터 > 주가 데이터 분석 모델
뉴스 데이터 > 뉴스 분석 모델

주가 데이터 분석 모델 + 뉴스 분석 모델 > 주식 시장 예측

```python
#날짜 타임으로 바꾸는 함수
reddit['Date'] = pd.to_datetime(reddit['Date'])
stock['Date'] = pd.to_datetime(stock['Date'])
```

```python
stock = stock.set_index('Date') #한번만 실행
```

```python
reddit_agre = reddit.groupby('Date').sum() # 같은 날짜의 뉴스 붙이기
```

```python
stock_new = pd.concat((stock, reddit_agre), axis=1)
stock_new.shape
```

```python
stock_new = stock_new.dropna() #  누락제거
stock_new.head(1)
```

```python
reddit_1d = stock_new['News'].values.reshape([-1]) # 1차원으로 reshape해서 어휘사전 만들기
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
vect = CountVectorizer().fit(reddit_1d)
len(vect.vocabulary_)
```

```python
reddit_sparse = vect.transform(reddit_1d)
```

```python
x_stock = stock_new.drop(columns='News').values # dense matrix
x_news = reddit_sparse.toarray() # sparse > dense
X = np.concatenate((x_stock, x_news), axis=1)
y = stock_new['Close'].values
X.shape, y.shape
```

```python
# 전날 X로 다음날 y 예측하기
X = X[:-1] # 최신 데이터 제거
y = y[1:]   # 가장 오래된 데이터 제거
```

```python
train_m = int(len(X)*0.75) # 과거 데이터 75%로 학습
X_train_no_new = X[:train_m, :6]
X_test_no_new = X[train_m:, :6]
X_train = X[:train_m]
X_test = X[train_m:]
y_train = y[:train_m]
y_test = y[train_m:]
```

```python
# 분류 모델로 fit후 score 측정
from sklearn.linear_model import Ridge #예측 Regression
model = Ridge().fit(X_train, y_train)
model.score(X_train, y_train), model.score(X_test, y_test) #결정계수
```
