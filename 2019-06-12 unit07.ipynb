```python
import csv # 모듈 호출

f = open('age.csv', 'r', encoding='utf-8')
data = csv.reader(f)

print(data)
result = []
print("인구 구조가 알고 싶은 지역의 이름을 입력하세요")
name = input()
for row in data :
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i))
print(result)

print("")
print(name, " 지역의 인구 구조")
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
