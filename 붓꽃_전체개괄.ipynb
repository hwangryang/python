```python
#p.44
from sklearn.model_selection import train_test_split #함수
```

```python
# 데이터를 학습용 75 % / 테스트용 25%를 나눠서 섞습니다(램덤하게).
x_train, x_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state=0)
# random_state=0 : 1회만 섞음(고정됨)
iris_dataset.data.shape, iris_dataset.target.shape
```

```python
#47 k-최근접 이웃 모델
from sklearn.neighbors import KNeighborsClassifier #대문자 시작시 클래스, 소문자 시작시 함수
knn = KNeighborsClassifier(n_neighbors=1)
# 분류기 객체 생성
```

```python
knn
```

```python
#p.48 훈련 데이터셋으로 모델을 만드는 .fit 함수
knn.fit(x_train, y_train)
# fit - 모델을 훈련 데이터에 맞게 핏팅 시켜라
```

```python
#p.48 예측하기
x_new = np.array([[5, 2.9, 1, 0.2]]) # 야생에서 꽃을 발견
# ['꽃받침 길이', '꽃받침 너비', '꽃잎 길이', '꽃잎 너비']
x_new.shape
```
